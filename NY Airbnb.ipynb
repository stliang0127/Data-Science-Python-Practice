```python
# prepare for prediction
df=pd.read_csv("AB_NYC_2019.csv")
```

```python
from sklearn import preprocessing
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
df1=df
df1.isnull().sum()
# 4 variables got null and need to deal with
```

```python
df1.drop(['name', 'host_name', 'last_review'],axis=1, inplace=True)
```

```python
df1.isnull().sum()
```

```python
df1['reviews_per_month'].fillna(0, inplace=True)
```

```python
df1.isnull().sum()
```

```python
# Code those categorical variables
le=preprocessing.LabelEncoder()
le.fit(df1['neighbourhood_group'])
```

```python
le.classes_
```

```python
df1['neighbourhood_group']=le.transform(df1['neighbourhood_group'])
```

```python
df1.head()
```

```python
le.fit(df1['neighbourhood'])
```

```python
le.classes_
```

```python
df1['neighbourhood']=le.transform(df1['neighbourhood'])
```

```python
le.fit(df1['room_type'])
le.classes_
```

```python
df1['room_type']=le.transform(df1['room_type'])
```

```python
# all the variables now turn to be numerical
df1.sort_values(by='price', ascending=True, inplace=True)
df1.head(10)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, LogisticRegression
```

```python
# Define the model
lm=LinearRegression()
```

```python
X=df1[['neighbourhood_group', 'neighbourhood', 'room_type', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'calculated_host_listings_count', 'availability_365']]
y=df1['price']
```

```python
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.2, random_state=123)
```

```python
lm.fit(X_train, y_train)
```

```python
predict=lm.predict(X_test)
```

```python
from sklearn import metrics
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error
```

```python
print("Rooted mean squared error is:")
np.sqrt(metrics.mean_squared_error(y_test, predict))
```

```python
print("R2 score is:")
r2=r2_score(y_test, predict)
r2*100
```

```python
print("Mean absolute error is:")
mean_absolute_error(y_test, predict)
```
